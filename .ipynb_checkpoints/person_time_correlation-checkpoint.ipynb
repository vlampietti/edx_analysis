```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
%matplotlib inline
```

```python
chapter_grades = pd.read_csv('sample_data/data_chapter_grades.csv')
person_course = pd.read_csv('sample_data/data_person_course.csv')
person_course_day = pd.read_csv('sample_data/data_person_course_day.csv')
problem_grades = pd.read_csv('sample_data/data_problem_grades.csv')
show_ans = pd.read_csv('sample_data/data_show_ans_stat_by_user.csv')
time_on_task = pd.read_csv('sample_data/data_time_on_task.csv')
time_on_task_total = pd.read_csv('sample_data/data_time_on_task_total.csv')
video_axis = pd.read_csv('sample_data/data_video_axis.csv')
video_stats = pd.read_csv('sample_data/data_video_stats.csv')
video_stats_day = pd.read_csv('sample_data/data_video_stats_day.csv')
```

```python
def display(columns, df):
    df.reset_index().plot(x='index',y=columns, figsize=(15, 7))
```

```python
total = person_course_day.groupby('date').sum().reset_index()
display(['nvideo','nproblems_attempted'],total)
#total.plot(x='date',y=['nvideo','nproblems_attempted'])
#total.plot(x='date',y='sum_dt')
```

```python
strip_cg = chapter_grades.loc[chapter_grades['chapter_mid']=='MITx/8.01r_3/chapter/exam:examfinal'].set_index('user_id')['chgrade']
top10per = list(strip_cg.sort_values(ascending=False).nlargest(59).index.values) # strip_cg has 589 entries
```

```python
top10pcd = person_course_day.loc[person_course_day['username'].isin(top10per)].groupby('date').sum()
top10pcd.columns = ['top_' + col for col in list(top10pcd)]
bot90pcd = person_course_day.loc[-person_course_day['username'].isin(top10per)].groupby('date').sum() / 9
bot90pcd.columns = ['bot_' + col for col in list(bot90pcd)]
join = pd.concat([top10pcd,bot90pcd], axis=1)
display(['top_nproblems_answered','bot_nproblems_answered'],join)
#make bin size function
```
