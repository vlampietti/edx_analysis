```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
%matplotlib inline
```

```python
chapter_grades = pd.read_csv('sample_data/data_chapter_grades.csv')
person_course = pd.read_csv('sample_data/data_person_course.csv')
person_course_day = pd.read_csv('sample_data/data_person_course_day.csv')
problem_grades = pd.read_csv('sample_data/data_problem_grades.csv')
show_ans = pd.read_csv('sample_data/data_show_ans_stat_by_user.csv')
time_on_task = pd.read_csv('sample_data/data_time_on_task.csv')
time_on_task_total = pd.read_csv('sample_data/data_time_on_task_total.csv')
video_axis = pd.read_csv('sample_data/data_video_axis.csv')
video_stats = pd.read_csv('sample_data/data_video_stats.csv')
video_stats_day = pd.read_csv('sample_data/data_video_stats_day.csv')
```

```python
def display_corr(columns, df):
    cols = columns
    if (columns is None):
        cols = list(df)
    
    fig = plt.figure(figsize=(9,9))
    ax = fig.add_subplot(111)
    cax = ax.matshow(df[cols].corr(), interpolation='nearest')
    fig.colorbar(cax)

    ax.set_xticklabels(['']+cols)
    ax.set_yticklabels(['']+cols)
    
    ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
    ax.yaxis.set_major_locator(ticker.MultipleLocator(1))
    
    plt.show()
```

```python
columns = ['nevents','ndays_act','nplay_video','nchapters','nprogcheck','nproblem_check','nshow_answer','nvideo','nvideos_unique_viewed','nvideos_total_watched','nseq_goto','nseek_video','npause_video','sum_dt']
display_corr(columns, person_course)
```

```python
show_ans = show_ans.assign(pertopar=show_ans['n_perfect']/show_ans['n_partial'])
show_ans = show_ans.assign(partoatt=show_ans['n_partial']/show_ans['n_attempted'])

columns = ['pct_show_answer_problem_seen','pct_show_answer_not_attempted','pct_show_answer_attempted','pct_show_answer_perfect','pct_show_answer_partial','pertopar','partoatt']
display_corr(columns=columns,df=show_ans)
```

```python
time_on_task_total['pervid'] = time_on_task_total['total_video_time_5'] / time_on_task_total['total_time_5']
time_on_task_total['perprob'] = time_on_task_total['total_problem_time_5'] / time_on_task_total['total_time_5']
time_on_task_total['pertxt'] = time_on_task_total['total_text_time_5'] / time_on_task_total['total_time_5']

columns = ['total_time_5','total_time_30','total_video_time_5','total_video_time_30','total_problem_time_5','total_problem_time_30','total_text_time_5','total_text_time_30','pervid','perprob','pertxt']
display_corr(columns=columns,df=time_on_task_total)
```

```python
strip_cg = chapter_grades.loc[chapter_grades['chapter_mid']=='MITx/8.01r_3/chapter/exam:examfinal'].set_index('user_id')['chgrade']
strip_pc = person_course.drop(['start_time','first_event','last_event','avg_dt','sdv_dt','max_dt','n_dt'],axis=1)
strip_sa = show_ans.drop(['explored','certified','verified'],axis=1)
strip_ttt = time_on_task_total.drop(['serial_video_time_30','serial_problem_time_30','serial_text_time_30'],axis=1)
join = pd.concat([strip_pc.set_index('user_id'),strip_sa.set_index('user_id'),strip_cg,strip_ttt.set_index('user_id')],axis=1)
join
display_corr(None,df=join)
```
